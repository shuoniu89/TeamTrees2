```python
import csv
import pandas as pd
from collections import defaultdict
import math
# merge the two csv files first 
enbang = pd.read_csv('./KEY/rejected_url_Enbang.csv')
cat = pd.read_csv('./KEY/rejected_url_Cat.csv')
merged = enbang.merge(cat, on='VIDEO_URL')
merged
```

```python
# drop columns we don't need
origin = pd.read_csv('./KEY/rejected_data.csv')
# print(origin.head(10))
# origin = origin.drop(['TASK_ID','SAME_TASK_GROUP_ID','STATUS','USER_ID','IP_ADDRESS','COUNTRY','START_TIME',
#            'FINISH_TIME','TIME_TOOK_TO_COMPLETE_TASK','event_name','speak_language','mrbeast','arborday','teamtrees_org',
#                     'noteble_youtuber','noteble_entrepreneur','quality','creativity','awareness'], axis = 1)
li1=[]
li2=[]
li3=[]
for d in merged["VIDEO_URL"]:
    cate_df=origin[origin.VIDEO_URL==d]
    tags=cate_df["category"].tolist()
    li1.append(tags[0])
    li2.append(tags[1])
    li3.append(tags[2])
merged["mw1"]=pd.Series(li1)
merged["mw2"]=pd.Series(li2)
merged["mw3"]=pd.Series(li3)
print(merged.head())

df = pd.DataFrame(merged, columns= ['VIDEO_URL', 'enbang_tag','Cat_tag','mw1','mw2','mw3'])

df.to_csv ('./KEY/retagged_data.csv', index = False, header=True) 
```
