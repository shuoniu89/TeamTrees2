```python
import json
import datetime
import dateutil
import isodate as isodate
import pytz
import math
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import csv
from datetime import timedelta
import dateutil.parser
import random
from langdetect import detect
from collections import defaultdict

youtuber_list_path="subscriber_network.json"

data=defaultdict(list)


def save_csv_file(data, file_name):
    data=[{"video_url":v["video_url"]} for v in data]
    chunks=[data[i:i + 50] for i in range(0, len(data), 50)]
    for i in range(len(chunks)):
        with open("microworkers/{}{}.csv".format(file_name, i), mode='w+', encoding="utf-8", newline='') as fp:
            csv_writer = csv.writer(fp, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
            first = True
            for v in chunks[i]:
                if first:
                    csv_writer.writerow([key for key, value in v.items()])
                    first = False
                csv_writer.writerow([value for key, value in v.items()])
                
english_country_code=["AG", "AU", "BS","BB","BZ","CA", "DM", "GD", "GY", "IE", "JM", "NZ", "KN", "LC", "VC", "TT", "GB", "US", "PH" "unknown"]

processed_list=set()

with open("CSV1.csv", 'r') as fp:
    spamreader = csv.reader(fp)
    counter=0
    for row in spamreader:
        if counter!=0:
            processed_list.add(row[9])
        counter+=1

print(processed_list)
with open(youtuber_list_path,'r') as fp:
    youtubers=json.loads(fp.read())["nodes"]
    for you in youtubers:
        for video in you["video_list"]:
            url="https://www.youtube.com/embed/"+video["video_id"]+"?autoplay=0"
            if url not in processed_list:
                data[video["language"]].append({
                    "channel_name":you["channel_name"],
                    "video_id":video["video_id"],
                    "language":video["language"],
                    "country":you["country"],
                    "title":video["title"],
                    "video_length":video["video_length"],
                    "video_url":url,
                })
                

counter=0
for k,v in data.items():
    counter+=len(v)
print(counter)

for key,item in data.items():
    print("key: {}, count:{}".format(key, len(item)))
    save_csv_file(item, key)


# save_csv_file([{"video_url":v["video_url"]} for v in data_no_en], "mc_none_en")



```
