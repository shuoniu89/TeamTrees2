```python
from matplotlib import pyplot as plt
from matplotlib import axes
from matplotlib import pyplot
import pandas as pd
import numpy as np
import math
from scipy import stats
import csv
import statistics

from collections import Counter
from collections import defaultdict

data=[]
result_path = "./KEY/"

with open(result_path + "final_meta_all_processed - no other.csv", "r+", encoding="utf-8") as fp:
    csv_reader=csv.DictReader(fp)
    for row in csv_reader:
        data.append(row)
```

```python
criteria = ["quality", "rel_vw", "rel_like"] #criteria to search for
ref = [("Creation","CREAT"),("Participation","PART"),
      ("Connection_comment","CXN-SP"), ("Connection_game","CXN-GAM")]

old_ref, new_ref = zip(*ref)

def convert_merged_tag(tagging):
    converted = "ISSUE"
    for i in ref:
        if i[0] == tagging:
            converted = i[1]
            break
    return converted        

celeb, pro, ama = [],[],[]


def label_diff(x1,x2,y,text):
    x = (x1+x2)/2
    y = 1.1*y
    dx = abs(x2-x1)

    props = {'connectionstyle':'bar','arrowstyle':'-',\
                 'shrinkA':20,'shrinkB':20,'linewidth':2}
    ax.annotate(text, xy=(x1,y+7), zorder=10)
    ax.annotate('', xy=(x1,y), xytext=(x2,y), arrowprops=props)

for d in data:
    list_crit = [convert_merged_tag(d["merged_tagging"])]
    if d["status"] == "celebrity":
        for crit in criteria:
            list_crit.append(d[crit])
        celeb.append(list_crit)
    if d["status"] == "professional":
        for crit in criteria:
            list_crit.append(d[crit])
        pro.append(list_crit)
    if d["status"] == "amateur":
        for crit in criteria:
            list_crit.append(d[crit])
        ama.append(list_crit)


i = 0  
rt_dict = defaultdict(list)
while i < len(criteria):
    for tag in new_ref:
        temp_dict = {
            "tag": tag,
            "celebrity": statistics.mean([float(d[i+1]) for d in celeb if d[0]==tag]),
            "professional": statistics.mean([float(d[i+1]) for d in pro if d[0]==tag]),
            "amateur": statistics.mean([float(d[i+1]) for d in ama if d[0]==tag])
        }
        rt_dict[criteria[i]].append(temp_dict)
    i += 1
        

x_val = ["celebrity", "professional", "amateur"]
x_val_num = range(len(x_val))
x_range = len(new_ref)

width = 0.2
plt.figure(figsize=(20,10))
i = 0

for crite in criteria:
    stderrs = []
    i+=1
    n = -1.5
    plt.subplot(2,3,i)
    for tag in new_ref:
        for dicts in rt_dict[crite]:
            if dicts["tag"] == tag:
                x,y = zip(*dicts.items()) 
                y_vals = y[1:len(x_val)+1]
                break
        stderr = stats.sem(y_vals, ddof = 1)
        stderrs.append(stderr)
        plt.bar([(x + n*(width)) for x in x_val_num], y_vals, width, label = tag, alpha = .7,
                 yerr = stderr, capsize = 3, error_kw = {"elinewidth":.9})
        n += 1
    scale = 1.5
    min_y = 0
    if stderrs: max_y = scale*math.ceil(max(stderrs) + max(y_vals))
    else: max_y = scale*math.ceil(max(y_vals))
    if crite == "rel_vw":
        crite = "Relative view"
        newscale = 3
        max_y = 7
        y_axis="Ratio"
    elif crite == "rel_like":
        crite = "Relative likes"
        newscale = 5
        max_y = 15
        y_axis="Like count / 100 views"
    elif crite == "quality":
        crite = "Quality rating"
        min_y = 2
        max_y = 5
        y_axis="Rating 1(very low)-5(very high)"
    plt.title(crite)
    plt.ylim(min_y, max_y)
    plt.ylabel(y_axis, fontsize=12)
    plt.xticks(x_val_num, x_val, fontsize=12)   
    plt.legend(loc="upper right")
    # Call the function
    
    
plt.savefig(result_path + "viz_qual_rt.png")
```
