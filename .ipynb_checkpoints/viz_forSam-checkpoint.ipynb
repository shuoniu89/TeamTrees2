```python
import pandas as pd
from matplotlib import pyplot as plt
from collections import Counter
import numpy as np

df2 = pd.read_csv("YouTube_CSV/channel_FINAL_merged_teamtrees.csv")
# df2.subscriberCount = _numeric(df2.subscriberCount, errors='coerce')
df2 = df2.loc[(df2["subscriberCount"] != -1)]
# df2["subscriberCount"] = df2["subscriberCount"].astype(str).astype(int)
df2["subscriberCount"] = sorted(df2["subscriberCount"])

# sorted_df = df2["subscriberCount"].sort_values(by = ["subscriberCount"], ascending = True)
# sorted_df.head()
subCount = df2["subscriberCount"]
plt.figure(figsize=(5,20))
# y-axis in histogram shows how many values/videos are in between each amount of values
bins = np.arange(0, 108000000, 100000)
plt.hist(subCount, bins = bins, edgecolor = "black",histtype='step')
plt.plot()
plt.xlabel("subscriberCount")
plt.ylabel("videoCount")
plt.xticks(np.arange(0, 108000000+1, 20.0))
```
