```python
from matplotlib import pyplot as plt
from matplotlib import axes
from matplotlib import pyplot
import pandas as pd
import numpy as np
import csv
from collections import Counter
from collections import defaultdict

data=[]
result_path = "./KEY/"

with open(result_path + "final_meta_all_processed - no other.csv", "r+", encoding="utf-8") as fp:
    csv_reader=csv.DictReader(fp)
    for row in csv_reader:
        data.append(row)

status = [d["status"] for d in data]
count = Counter(status)
labels = []
vals = []
total = len(status)

for k,v in count.items():
    labels.append(k)
    vals.append(v)
    
def make_autopct(values):
    def my_autopct(pct):
        val = int(round(pct*total/100.0))
        return '{p:.2f}%\n {v:d} videos'.format(p=pct,v=val)
    return my_autopct
    
```

```python
def mk_groups(data):
    try:
        newdata = data.items()
    except:
        return

    thisgroup = []
    groups = []
    for key, value in newdata:
        newgroups = mk_groups(value)
        if newgroups is None:
            thisgroup.append((key, value))
        else:
            thisgroup.append((key, len(newgroups[-1])))
            if groups:
                groups = [g + n for n, g in zip(newgroups, groups)]
            else:
                groups = newgroups
    return [thisgroup] + groups

def add_line(ax, xpos, ypos):
    line = plt.Line2D([xpos, xpos], [ypos - 0.5, ypos+.1],
                      transform=ax.transAxes, color='black')
    line.set_clip_on(False)
    ax.add_line(line)

def label_group_bar(ax, data):
    groups = mk_groups(data)
    xy = groups.pop()
    x, y = zip(*xy)
    ly = len(y)

    #draw the bar
    cat, num = zip(*groups[0])
    j = 1
    k = 0
    for i in num:
        xticks = range(j, j + i) #num is list of how many items in each big category.
                                #So xticks goes from, say, 1-3, 3-5 if num is [2,2]
        j += i
        bars = ax.bar(xticks, y[k:k+i],align='center') #y is the list of all y values to be plotted. same reasoning w xticks
        k += i
        for col in bars:
            height = col.get_height()
            plt.text(col.get_x() + col.get_width()/2.0, height, '%d' % int(height), ha='center', va='bottom')
     
    xticks = range(1, ly + 1)
    ax.set_xticks(xticks)
    ax.set_xlim(0.4, ly + .5)
    ax.set_xticklabels(x, rotation=90, ha='right')
    scale = 1. / ly
    for pos in range(ly + 1):
        add_line(ax, pos * scale, -.1)
    ypos = -.2
    while groups:
        group = groups.pop()
        pos = 0
        for label, rpos in group:
            lxpos = (pos + 0.5 * rpos) * scale
            ax.text(lxpos, ypos-0.47, label, ha='center', transform=ax.transAxes)
            add_line(ax, pos * scale, ypos)
            pos += rpos
        add_line(ax, pos * scale, ypos)
        ypos -= .1

ref = {"CREAT":
                   ["create object","performance","knowledge"],
            "PART":
                   ["plant trees","donating"],
            "CXN_SP":
                   ["spread word","comment news", "comment video", "livestream stat"],
            "CXN\nGAM":
                   ["livestream game"],
           "Other":
                   ["other"]
           }
head_ref, item_ref = zip(*ref.items())
item_ref = [i for sublist in item_ref for i in sublist]


def title_convert(title):
    converted = "other"
    for i in item_ref:
        if (title[-4:] == i[-4:]):
            converted = i
            break
    return converted

if __name__ == '__main__':
    

    data = []
    with open(result_path + "final_accepted_taggings.csv", "r+", encoding="utf-8") as fp:
        csv_reader=csv.DictReader(fp)
        for row in csv_reader:
            data.append(row)
    tag_list = [title_convert(d["tagging"]) for d in data]
    tag_count = Counter(tag_list)
    data = {}
    
    for head in head_ref:
        temp_dict = {}
        for k,v in tag_count.items():
            if (k in ref[head]):
                temp_dict[k] = v              
        data[head] = temp_dict
                  
#     data = {"CREAT":
#                    {"create object": 10,
#                     "performance": 60,
#                     "knowledge":12},
#             "PART":
#                    {"plant trees": 5,
#                     "donating": 6}
#             ,
#             "CXN_SP":
#                    {"spread word": 4,
#                     "comment news": 7,
#                     "comment video":5,
#                     "livestream stat": 6},
#             "CXN\nGAM":
#                    {"livestream game": 2},
#             "Other":
#                     {"other":50}
#            }
    fig = plt.figure(figsize=(15,8))
    ax = fig.add_subplot(1,2,1)
    label_group_bar(ax, data)
    plt.ylabel("Number of Videos")
    plt.ylim(0,120)
    fig.subplots_adjust(bottom=0.5)
    
    plt.subplot(1,2,2)
    wedges, texts, autotexts = plt.pie(vals, autopct=make_autopct(vals),
                                    shadow=True, startangle=90)

    plt.legend(wedges, labels,
          title="Status",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

    plt.title("Number and Percentage by Different YouTubers")

    plt.savefig(result_path + "final_graph.png")
```
