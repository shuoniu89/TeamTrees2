```python
import csv
import os
import statistics
import math
from collections import defaultdict
from collections import Counter
import json

directory="./data_tagging_j8_origin" #use your own directory path (I saved all the csv in a separate folder called data)

    
def save_csv(file,data):
    with open("data_tagging_j8/"+file, mode='w', encoding="utf-8") as csv_result:
        csv_writer = csv.writer(csv_result,
                                delimiter = ",",
                                quotechar = '"',
                                quoting = csv.QUOTE_MINIMAL,
                                lineterminator = "\n")
        isFirst=True
        for i in data:
            if isFirst:
                csv_writer.writerow(i.keys())
                isFirst=False
            csv_writer.writerow(i.values())

wrong_data=[]
with open("KEY/broken_data.csv", encoding="utf-8") as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        wrong_data.append(row)
wrong_data=[d["video_url"] for d in wrong_data]

fixes=[]
with open("KEY/fix_wrong_1.csv", encoding="utf-8") as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        if row["STATUS"] !="NOK" and row["VIDEO_URL"] not in wrong_data:
            fixes.append(row)
        
with open("KEY/fix_wrong_2.csv", encoding="utf-8") as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        if row["STATUS"] !="NOK" and row["VIDEO_URL"] not in wrong_data:
            fixes.append(row)        

# print(json.dumps(fixes, indent=4))

               
for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        data = []
        with open(directory+"/"+filename, mode = "r+", encoding="utf-8") as csv_file:
            csv_reader = csv.DictReader(csv_file)
            for row in csv_reader:
                data.append(row)
        same_work_id={}
        n_list=[]
        for d in data:
            if d["STATUS"]=="OK" and d["event_name"]!="TeamTrees" and d["VIDEO_URL"] not in wrong_data:
                print(f"remove {filename} --> {d['VIDEO_URL']}")
                d["STATUS"]="NOK"
                replace={}
                for f in fixes:
                    if f["VIDEO_URL"]==d["VIDEO_URL"]:
                      replace=f
                      replace["SAME_TASK_GROUP_ID"]=d["SAME_TASK_GROUP_ID"]
#                       replace["SAME_TASK_GROUP_ID"]="fixed"
                      replace={k:replace[k] for k in d.keys()}
                      break
#                 assert replace != {}
#                 assert replace["STATUS"] == "OK"
                fixes.remove(replace)
                n_list.append(replace)
        data.extend(n_list)
        save_csv(filename, data)
    else:
        pass

```
